```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
import os
```

```python
filename = "Loan.csv"
df = pd.read_csv(filename)
print(df.shape)
df.head()
```

```python

#df.drop(0, inplace = True)
columns = df.columns
features = columns.drop('loan_status')
target = ['loan_status']
x = df[features]
y=df[target]
```

```python
print(features.size)
y.head()
```

```python
x = x.astype(float)
y = y.astype(float)
#train_x, test_x, train_y, test_y = train_test_split(x,y,test_size = 0.2, random_state=0)

train = pd.read_csv("../src/resources/datasets/Loan/Loan.train5.csv")
test = pd.read_csv("../src/resources/datasets/Loan/Loan.test5.csv")
train_x = train[features]
train_y = train[target]
test_x = train[features]
test_y = train[target]
```

```python
train_y.head()
```

```python
n_estimators = 10
max_depth = 3
myclassifier = RandomForestClassifier(random_state=42, criterion = 'gini', n_estimators = n_estimators, max_features = 'sqrt', max_depth =max_depth)
params = {}
n_fold = 5

gscv_rf = GridSearchCV(myclassifier, param_grid = params, scoring='recall', cv=n_fold, refit=True)

gscv_rf.fit(train_x, train_y.loan_status)
```

```python
rf = gscv_rf.best_estimator_
trees = rf.estimators_
for tree in trees:
    print(tree.tree_.node_count)
```

```python
print(trees[1].tree_.feature[3])
print(trees[1].tree_.value)
print(trees[1].tree_.value[0])
print(trees[1].tree_.impurity[0])
print(trees[1].tree_.value[14])

```

```python
firstTree = trees[0].tree_
n_nodes = firstTree.node_count
children_left = firstTree.children_left
children_right = firstTree.children_right
feature = firstTree.feature
threshold = firstTree.threshold

node_depth = np.zeros(shape=n_nodes, dtype=np.int64)
is_leaves = np.zeros(shape=n_nodes, dtype=bool)
stack = [(0, 0)]  # start with the root node id (0) and its depth (0)
while len(stack) > 0:
    # `pop` ensures each node is only visited once
    node_id, depth = stack.pop()
    node_depth[node_id] = depth

    # If the left and right child of a node is not the same we have a split
    # node
    is_split_node = children_left[node_id] != children_right[node_id]
    # If a split node, append left and right children and depth to `stack`
    # so we can loop through them
    if is_split_node:
        stack.append((children_left[node_id], depth + 1))
        stack.append((children_right[node_id], depth + 1))
    else:
        is_leaves[node_id] = True

print(
    "The binary tree structure has {n} nodes and has "
    "the following tree structure:\n".format(n=n_nodes)
)
for i in range(n_nodes):
    if is_leaves[i]:
        print(firstTree.value[i])
        print(firstTree.impurity[i])
        
    else:
        print("")
        
```

```python
f = open(str(absolutePath), 'w')
f.write("DATASET_NAME: " + filename+"\n")
f.write("ENSEMBLE: RF\n")
f.write("NB_TREES: " + str(n_estimators)+"\n")
f.write("NB_FEATURES: " + str(features.size)+"\n")
f.write("NB_CLASSES: 2\n")
f.write("MAX_TREE_DEPTH: " + str(max_depth)+"\n")
f.write("Format: node / node type (LN - leave node, IN - internal node) left child / right child / feature / threshold / node_depth / majority class (starts with index 0)\n")

for j in range(0, len(trees)):
    f.write("\n[TREE " + str(j)+"]\n")
    tree = trees[j].tree_
    n_nodes = tree.node_count
    f.write("NB_NODES: " + str(n_nodes)+"\n")

    children_left = tree.children_left
    children_right = tree.children_right
    feature = tree.feature
    threshold = tree.threshold
    value = tree.value

    node_depth = np.zeros(shape=n_nodes, dtype=np.int64)
    is_leaves = np.zeros(shape=n_nodes, dtype=bool)
    stack = [(0, 0)]  # start with the root node id (0) and its depth (0)
    while len(stack) > 0:
        # `pop` ensures each node is only visited once
        node_id, depth = stack.pop()
        node_depth[node_id] = depth

        # If the left and right child of a node is not the same we have a split
        # node
        is_split_node = children_left[node_id] != children_right[node_id]
        # If a split node, append left and right children and depth to `stack`
        # so we can loop through them
        if is_split_node:
            stack.append((children_left[node_id], depth + 1))
            stack.append((children_right[node_id], depth + 1))
        else:
            is_leaves[node_id] = True

    for i in range(n_nodes):
        f.write(str(i) + " ")
        if is_leaves[i]:
            f.write("LN ")
            f.write(str(children_left[i])+ " " + str(children_right[i]) + " " + str(feature[i]) + " " + str(threshold[i]) + " " + str(node_depth[i]) + " ")
            if(value[i][0][0]>=value[i][0][1]):
                f.write("0")
            else:
                f.write("1")
            
        else:
            f.write("IN ")
            f.write(str(children_left[i])+ " " + str(children_right[i]) + " " + str(feature[i]) + " " + str(threshold[i]) + " " + str(node_depth[i]) + " ")
            f.write("-1")
        f.write("\n")
f.write("\n")
f.flush()
f.close()
    
    
```
